```python
class onehot:
    def __init__(self):
        self.encoder={}
        self.decoder={}
        self.n_cats=0
    def encode(self,col):
        cats=np.unique(col)
        self.n_cats=len(cats)
        n=0
        for i in cats:
            self.encoder[i]=n
            self.decoder[n]=i
            n+=1
        
        output=[]
        for i in col:
            code=[0]*self.n_cats
            code[self.encoder[i]]=1
            output.append(code)
        return np.array(output)
    def decode(self,col):
        output=[]
        card=np.arange(self.n_cats)
        for i in col:
            n=i@card.T
            output.append(self.decoder[n])       
        return output
```
