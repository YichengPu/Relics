```python
class onehot:
    def __init__(self):
        self.encoder={}
        self.decoder={}
        self.n_cats=0
    def encode(self,col):
        cats=np.unique(col)
        self.n_cats=len(cats)
        n=0
        for i in cats:
            self.encoder[i]=n
            self.decoder[n]=i
            n+=1
        
        output=[]
        for i in col:
            code=[0]*self.n_cats
            code[self.encoder[i]]=1
            output.append(code)
        return np.array(output)
    def decode(self,col):
        output=[]
        card=np.arange(self.n_cats)
        for i in col:
            n=i@card.T
            output.append(self.decoder[n])       
        return output
```

```python
def g_log(x,epsilon=1e-7):
    return np.log(x+epsilon)

def entropy(arr,split,ep=1e-7):
    p=sum(arr[arr>=split])/len(arr)
    return -(p*g_log(p+ep)+(1-p)*g_log(1-p+ep))

def jd_entropy(arr1,split1,arr2,split2):
    
    '''
    joint distribution entropy
    '''
    
    arr1=np.array(arr1)
    arr2=np.array(arr2)
    p1=sum((arr1>=split1) & (arr2>=split2))/len(arr1)
    p2=sum((arr1<split1) & (arr2>=split2))/len(arr1)
    p3=sum((arr1>=split1) & (arr2<split2))/len(arr1)
    p4=sum((arr1<split1) & (arr2<split2))/len(arr1)
    
    return -(p1*g_log(p1)+p2*g_log(p2)+p3*g_log(p3)+p4*g_log(p4))


def split(mat,col,v):
    left=[]
    right=[]
    
    for i in range(mat.shape[0]):
        if mat[i,col] <= v:
            left.append(i)
        else:
            right.append(i)
        
        
    return (left,right)

def grow(mat,target):
    
    nrow,ncol=mat.shape
    
    
    best_split=(-1,-1)
    best_gain=0
    
    for i in range(nrow):
        for j in range(ncol):
            left,right=split(mat,j,mat[i][j])
            
            leng_left=len(left)
            leng_right=len(right)
#             print(i,j,left,right)
            if leng_left>0 and leng_right>0:
                
            
                info_gain=entropy(target,1)-len(left)/nrow*entropy(target[left],1)\
                    -len(right)/nrow*entropy(target[right],1)
                    
#                 print(entropy(target,1),len(left)/nrow*entropy(target[left],1)\
#                     ,len(right)/nrow*entropy(target[right],1))

                print(info_gain)
                if info_gain> best_gain:
                    best_split=(i,j)
                    best_gain=info_gain
    
    left,right=split(mat,best_split[1],mat[best_split])
    return best_split,mat[left,:],mat[right,:],target[left],target[right] 
    
class decision_tree:
    def __init__(self):
        self.tree={}
        
        return

    def fit(self,x,y):
        '''
        x: array of size(n,p)
        y: binary arrary of size(n)
        '''
        #one split

        n=x.shape[0]
        p=x.shape[1]
        for i in range(p):
            feature=x[:,feature]
            for d_point in feature:
                H_attr=entropy(feature,d_point)
                H_DA=0
        return
    def predict(self):
        return
```
